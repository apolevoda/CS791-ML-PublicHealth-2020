```python
import plotly.graph_objects as go
import pandas as pd
```

```python
data = pd.read_csv('FLDExtract.csv', error_bad_lines=False)
cities = {}
for i in data["Employer City"]:
    pass
    if i in cities:
        cities[i] += 1
    else:
        cities[i] = 1

cities = {k: v for k, v in sorted(cities.items(), key=lambda item: item[1], reverse=True)}

colors = ["royalblue", "crimson", "lightseagreen", "orange", "lightgrey"]
limits = [(150,500), (100,150), (50,100), (20,50), (0,20)]
scale = 5000

cities_lst = []
for k,v in cities.items():
    cities_lst.append(k)

r_lon = []
r_lat = []
r_cities = []
r_sizes = []
r_colors = []
        
        
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv')
for i in cities_lst:
    if (type(i) == str):
        cur = df.loc[df["name"] == i+" "]
        if not cur.empty:
            r_cities.append(i)
            r_lon.append(cur["lon"].values[0])
            r_lat.append(cur["lat"].values[0])
            r_sizes.append(cities[i])
# print(r_cities)
ind = []
for i in range(len(r_cities)):
    ind.append(i)

lon = pd.Series(r_lon, index = ind) 
lat = pd.Series(r_lat, index = ind)
text = pd.Series(r_cities, index=ind)
size = pd.Series(r_sizes, index=ind)
# print(lat)

for i in r_cities:
    if cities[i] > 150:
        r_colors.append(4)
    elif cities[i] > 100:
        r_colors.append(3)
    elif cities[i] > 50:
        r_colors.append(2)
    elif cities[i] > 20:
        r_colors.append(1)
    else:
        r_colors.append(0)
# print(r_colors)


fig = go.Figure()


record_indices = [0]
cur_color = 4
for i in range(len(r_colors)):
    if cur_color != r_colors[i]:
        record_indices.append(i)
        cur_color = r_colors[i]
# print(record_indices)
# print(r_colors[:25])
record_indices.append(len(r_cities))

formalized_lon = []
formalized_lat = []
formalized_text = []
formalized_size = []
# print(len(record_indices))
for i in range(len(record_indices)-1):
#     index = record[i]
    formalized_lon.append(lon[record_indices[i]:record_indices[i+1]])
    formalized_lat.append(lat[record_indices[i]:record_indices[i+1]])
    formalized_text.append(text[record_indices[i]:record_indices[i+1]])
    formalized_size.append(size[record_indices[i]:record_indices[i+1]])

# for i in range(4):
#     print(type(formalized_lon[i]))
for i in range(5):
    lim = limits[i]
    fig.add_trace(go.Scattergeo(
        locationmode = 'USA-states',
        lon=formalized_lon[i],
        lat=formalized_lat[i],
        text=formalized_text[i],
        marker = dict(
            size=formalized_size[i],
            color = colors[i],
            line_color='rgb(40,40,40)',
            line_width=0.5,
            sizemode = 'area'
#         )))
    ),
    name = '{0} - {1}'.format(lim[0],lim[1])))

# fig.add_trace(go.Scattergeo(
#     locationmode = 'USA-states',
#     lon=lon,
#     lat=lat,
#     text=text,
#     marker = dict(
#         size=size,
# #         color = colors[i],
#         line_color='rgb(40,40,40)',
#         line_width=0.5,
#         sizemode = 'area'
#     )))
    
fig.update_layout(
        title_text = '2015-2020 Complaints Bubble Chart(Click legent to toggle traces)',
        showlegend = True,
        geo = dict(
            scope = 'usa',
            landcolor = 'rgb(217, 217, 217)',
        )
    )
fig.show()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Data.csv', error_bad_lines=False,encoding= 'unicode_escape')

catagories = {}

for i in data["Violation Description "]:
    if i not in catagories:
        catagories[i] = 1
    else:
        catagories[i] += 1
catagories = {k: v for k, v in sorted(catagories.items(), key=lambda item: item[1], reverse=True)}

kinds = []
numbers = []
for k,v in catagories.items():
    if v > 10:
        kinds.append(k)
        numbers.append(v)
    
y_pos = np.arange(len(kinds))
short_rep = {}
for i in range(len(kinds)):
    short_rep[i] = kinds[i]

plt.bar(y_pos, numbers, align='center', alpha=0.5)
plt.xticks(y_pos, y_pos)
plt.ylabel('Number of enforcements')
plt.title('Catagories')
for k,v in short_rep.items():
    print(k," : ", v)
```

```python
data = pd.read_csv('Data.csv', error_bad_lines=False,encoding= 'unicode_escape')
```

```python
cities = {}

for i in data["City, State "]:
    pass
    try:
        a = i.split(",")
        if a[0] in cities:
            cities[a[0]] += 1
        else:
            cities[a[0]] = 1
    except:
        pass

cities = {k: v for k, v in sorted(cities.items(), key=lambda item: item[1], reverse=True)}

cities_lst = []
for k,v in cities.items():
    cities_lst.append(k)
    
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_us_cities.csv')

r_cities = []
r_sizes = []
r_pop = []
for i in cities_lst:
    if (type(i) == str):
        cur = df.loc[df["name"] == i+" "]
        if not cur.empty:
            r_cities.append(i)
            r_sizes.append(cities[i])
            r_pop.append(cur["pop"].values[0])

a = [r_sizes[i] / r_pop[i] for i in range(len(r_pop))] * 100
a = a[:15]
r_cities = r_cities[:len(a)]


y_pos = np.arange(len(a))
short_rep = {}

for i in range(len(a)):
    short_rep[i] = r_cities[i]
    
plt.bar(y_pos, a, align='center', alpha=0.5)
plt.xticks(y_pos, y_pos)
plt.ylabel('Percentage')
plt.title('Enforcements cases / Number of population')
for k,v in short_rep.items():
    print(k,v)
```
